### import libraries

```python
# Importing the numpy library for numerical computations
import numpy as np 

# Importing the pandas library for data manipulation and analysis
import pandas as pd 

# Importing the matplotlib library's pyplot module for data visualization
import matplotlib.pyplot as plt 

# Ensuring that matplotlib plots are displayed inline in the notebook
%matplotlib inline

# Importing the seaborn library for making attractive and informative statistical graphics
import seaborn as sns
```

```python
# Reading the CSV file named 'Diwali Sales Data.csv' into a pandas DataFrame
# The 'encoding' parameter is set to 'unicode_escape' to handle any special characters in the file
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
  
```

```python
# Displaying the first 5 rows of the DataFrame to get an initial overview of the data
df.head()
```

```python
# Displaying a concise summary of the DataFrame, including the number of non-null entries and data types of each column
df.info()
```

```python
# Dropping the columns 'Status' and 'unnamed1' from the DataFrame
# Setting inplace=True to modify the DataFrame directly
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)


#This specifies to drop columns named 'Status' and 'unnamed1'. The axis=1 parameter indicates that we are referring to columns.
#inplace=True: This modifies the DataFrame in place, meaning the changes are applied directly to df without returning a new DataFrame.
```

```python
# Checking for null values in each column of the DataFrame and summing them up
pd.isnull(df).sum()


#This command will return a Series where each column name is paired with the count of null values in that column. 
#If a column has no null values, its count will be zero.
```

```python
# Dropping rows with null values from the DataFrame
# Setting inplace=True to modify the DataFrame directly
df.dropna(inplace=True)

# This will remove any row where at least one element is missing (NaN). After running this command, 
#your DataFrame df will no longer contain rows with null values.
```

```python
# Changing the data type of the 'Amount' column to integer
df['Amount'] = df['Amount'].astype('int')


#Explanation:

#df['Amount']: This selects the 'Amount' column in your DataFrame 
#df.astype('int'): This converts the data type of the selected column to integer. This assumes that the 'Amount' column originally contains 
#numerical values that can be safely converted to integers without loss of information.
#After running this command, the 'Amount' column in your DataFrame df will have its data type changed to integer (int). This can be useful when 
#you need to perform numerical operations or when the original data type of the column was incorrectly interpreted as something else.
```

```python
# Checking the data type of the 'Amount' column in the DataFrame
df['Amount'].dtypes
```

```python
# Displaying the list of column names in the DataFrame
df.columns
```

```python
# Renaming the column 'Marital_Status' to 'Shaadi' in the DataFrame
df.rename(columns={'Marital_Status': 'Shaadi'}, inplace=True)

#Explanation:

#df.rename(columns={'Marital_Status': 'Shaadi'}): This specifies that you want to rename the column named 'Marital_Status' to 'Shaadi'. 
#The columns parameter takes a dictionary where keys are the current column names and values are the new column names.
#inplace=True: This modifies the DataFrame directly. If inplace is not set to True, a new DataFrame with the renamed column will be returned 
#without modifying the original DataFrame df.
#After running this command, the column 'Marital_Status' in your DataFrame df will be renamed to 'Shaadi'
```

# Exploratory Data Analysis

##### Exploratory Data Analysis (EDA) is a critical first step in analyzing a dataset. It involves summarizing the main characteristics of the data, often with visual methods. Here’s a structured approach to performing EDA using Python libraries like Pandas, Matplotlib, and Seaborn:

## Gender

```python
# Plotting a bar chart for 'Gender' and its count using Seaborn's countplot
ax = sns.countplot(x='Gender', data=df)

# Adding labels to each bar in the bar chart
for bars in ax.containers:
    ax.bar_label(bars)



# Explanation:

# sns.countplot(x='Gender', data=df): This creates a bar chart with the x-axis representing the 'Gender' column and the y-axis showing the count of 
# each category in the 'Gender' column.
# ax.containers: This gets the bar containers (patches) for the bars in the plot.
# ax.bar_label(bars): This adds labels to each bar, displaying the count on top of the bars for better readability.
# This will create a bar chart where each bar represents the count of each gender category, with the count value labeled on top of each bar.
```

```python
# Grouping the data by 'Gender' and calculating the total 'Amount' for each gender
# Sorting the resulting DataFrame by 'Amount' in descending order
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting a bar chart using Seaborn's barplot to show 'Gender' vs. total 'Amount'
sns.barplot(x='Gender', y='Amount', data=sales_gen)

# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

## Age

```python
# Creating a count plot for 'Age Group' with 'Gender' as the hue
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

# Adding labels to each bar in the plot
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

#### Total Amount vs Age Group

```python
# Grouping the data by 'Age Group' and calculating the total 'Amount' for each age group
# Sorting the resulting DataFrame by 'Amount' in descending order
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting a bar chart using Seaborn's barplot to show 'Age Group' vs. total 'Amount'
sns.barplot(x='Age Group', y='Amount', data=sales_age)

# Adding labels to each bar in the bar chart
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age)
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

## State

#### total number of orders from top 10 states

```python
# Grouping the data by 'State' and calculating the total 'Orders' for each state
# Sorting the resulting DataFrame by 'Orders' in descending order and selecting the top 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(15,5)})

# Plotting a bar chart using Seaborn's barplot to show 'State' vs. total 'Orders' for the top 10 states
sns.barplot(data=sales_state, x='State', y='Orders')

# Adding labels to each bar in the bar chart
ax = sns.barplot(data=sales_state, x='State', y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

#### total amount/sales from top 10 states

```python
# Grouping the data by 'State' and calculating the total 'Amount' for each state
# Sorting the resulting DataFrame by 'Amount' in descending order and selecting the top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(15,5)})

# Plotting a bar chart using Seaborn's barplot to show 'State' vs. total 'Amount' for the top 10 states
sns.barplot(data=sales_state, x='State', y='Amount')

# Adding labels to each bar in the bar chart
ax = sns.barplot(data=sales_state, x='State', y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

## Marital Status

##### the code to create a count plot for 'Marital_Status'

```python
# Setting the size of the plot
sns.set(rc={'figure.figsize':(7,5)})

# Creating a count plot for 'Marital_Status'
ax = sns.countplot(data=df, x='Shaadi')

# Adding labels to each bar in the plot
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

##### to create a bar chart showing the total amount of sales by marital status and gender

```python
# Grouping the data by 'Marital_Status' and 'Gender', and calculating the total 'Amount' for each group
# Sorting the resulting DataFrame by 'Amount' in descending order
sales_state = df.groupby(['Shaadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(6,5)})

# Plotting a bar chart using Seaborn's barplot to show 'Marital_Status' vs. total 'Amount' with 'Gender' as hue
ax = sns.barplot(data=sales_state, x='Shaadi', y='Amount', hue='Gender')

# Adding labels to each bar in the bar chart
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

##### create a bar chart showing the total amount of sales by occupation

```python
# Grouping the data by 'Occupation' and calculating the total 'Amount' for each occupation
# Sorting the resulting DataFrame by 'Amount' in descending order
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(20,5)})

# Plotting a bar chart using Seaborn's barplot to show 'Occupation' vs. total 'Amount'
sns.barplot(data=sales_state, x='Occupation', y='Amount')

ax = sns.barplot(data=sales_state, x='Occupation', y='Amount')

# Adding labels to each bar in the plot
for bars in ax.containers:
    ax.bar_label(bars)
# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

#### create a count plot for 'Product_Category' with labels on each bar

```python
# Setting the size of the plot
sns.set(rc={'figure.figsize':(20,5)})

# Creating a count plot for 'Product_Category'
ax = sns.countplot(data=df, x='Product_Category')

# Adding labels to each bar in the plot
for bars in ax.containers:
    ax.bar_label(bars)

# Displaying the plot
plt.show()

## a count plot showing the number of occurrences for each product category ('Product_Category'), with the count values labeled on top of each bar
```

#### create a bar chart showing the total amount of sales for the top 10 product categories

```python
# Grouping the data by 'Product_Category' and calculating the total 'Amount' for each category
# Sorting the resulting DataFrame by 'Amount' in descending order and selecting the top 10 categories
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(20,5)})

# Plotting a bar chart using Seaborn's barplot to show 'Product_Category' vs. total 'Amount' for the top 10 categories
sns.barplot(data=sales_state, x='Product_Category', y='Amount')
ax = sns.barplot(data=sales_state, x='Product_Category', y='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
# Displaying the plot
plt.show()
```

###### From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

#### create a bar chart showing the total number of orders for the top 10 products based on their IDs, 

```python
# Grouping the data by 'Product_ID' and calculating the total 'Orders' for each product
# Sorting the resulting DataFrame by 'Orders' in descending order and selecting the top 10 products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Setting the size of the plot
sns.set(rc={'figure.figsize':(20,5)})

# Plotting a bar chart using Seaborn's barplot to show 'Product_ID' vs. total 'Orders' for the top 10 products
sns.barplot(data=sales_state, x='Product_ID', y='Orders')

ax = sns.barplot(data=sales_state, x='Product_ID', y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
# Displaying the plot
plt.show()
```

##### # top 10 most sold products (same thing as above)

```python
# Calculate the total number of orders for each product and select the top 10 most sold products
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Create a figure and axis (subplots) using matplotlib
fig1, ax1 = plt.subplots(figsize=(12, 7))

# Plotting a bar chart for the top 10 most sold products
top_products.plot(kind='bar', ax=ax1)

# Adding labels and title
ax1.set_xlabel('Product ID')
ax1.set_ylabel('Total Orders')
ax1.set_title('Top 10 Most Sold Products')

# Displaying the plot
plt.show()
```
